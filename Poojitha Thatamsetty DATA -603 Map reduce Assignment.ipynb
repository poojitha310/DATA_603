**EXTRACTION OF DATA**

```python
import PyPDF2
import os

pdf_path = r"C:\Users\pooji\Documents\Data_603\HarryPotter.pdf"

output_dir = r"C:\Users\pooji\Documents\Data_603"

os.makedirs(output_dir, exist_ok=True)

def extract_pages(pdf_path, start_page, end_page, output_filename):
    output_file = os.path.join(output_dir, output_filename)
    
    with open(pdf_path, 'rb') as pdf_file:
        reader = PyPDF2.PdfReader(pdf_file)
        text = ""
        for page_num in range(start_page - 1, end_page):  # Page numbers are 0-indexed
            page = reader.pages[page_num]
            text += page.extract_text() if page.extract_text() else ""

        # Save the extracted text to a file
        with open(output_file, 'w', encoding='utf-8') as output:
            output.write(text)

# Extract pages 31-41 for file1.txt
extract_pages(pdf_path, 31, 41, "file1.txt")

# Extract pages 100-110 for file2.txt
extract_pages(pdf_path, 100, 110, "file2.txt")

print(f"Pages extracted and saved to {output_dir}\file1.txt and {output_dir}\file2.txt.")
```

```python
import re
from itertools import groupby
import pandas as pd
```

```python
def mapfunction(m_var):
    cleaned = ''.join([i for i in m_var.lower()])
    cleaned = re.sub(r'[^\w\s]', '', cleaned)
    return [cleaned, 1]
```

```python
def reductionfunction(key, values):
    count = [x[1] for x in values]
    return [key, sum(count)]
```

```python
file1_path = r"C:\Users\pooji\Documents\Data_603\file1.txt"
file2_path = r"C:\Users\pooji\Documents\Data_603\file2.txt"
```

```python
with open(file1_path, encoding='utf8') as f:
    words = [word for line in f for word in line.split()]
```

```python
map_result = map(mapfunction, words)
map_result_sorted = sorted(map_result, key=lambda x: x[0])
reduce_result = []
for k, g in groupby(map_result_sorted, key=lambda x: x[0]):
    reduce_result.append(reductionfunction(k, list(g)))

print("Word counts from file1.txt:")
print(reduce_result)
```

```python
with open(file2_path, encoding='utf8') as h:
    words1 = [word for line in h for word in line.split()]
```

```python
second_map = map(mapfunction, words1)
second_sorted = sorted(second_map, key=lambda x: x[0])
reducer_sec = []
for l, m in groupby(second_sorted, key=lambda x: x[0]):
    reducer_sec.append(reductionfunction(l, list(m)))
```

```python
print(reducer_sec)
```
